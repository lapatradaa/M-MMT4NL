```python
# Corrected GPT-4o PromptOps data
categories = ["Taxonomy", "Negation", "Coreference", "SRL", "Fairness", "Robustness"]
x = np.arange(len(categories))

with_context_gpt = {
    "Zero-shot": [94, 98, 92, 94, 94, 93],
    "One-shot":  [92, 98, 96, 98, 92, 93],
    "Few-shot":  [98, 98, 98, 98, 92, 91],
}

without_context_gpt = {
    "Zero-shot": [86, 90, 98, 88, 82, 81],
    "One-shot":  [90, 92, 92, 92, 86, 78],
    "Few-shot":  [94, 92, 94, 90, 84, 83],
}

bar_width = 0.25

def plot_with_numbers(data, title):
    fig, ax = plt.subplots(figsize=(10, 6))
    offsets = [-bar_width, 0, bar_width]
    labels = ["Zero-shot", "One-shot", "Few-shot"]
    colors = ["orange", "skyblue", "seagreen"]

    for i, (label, offset, color) in enumerate(zip(labels, offsets, colors)):
        bars = ax.bar(x + offset, data[label], width=bar_width, label=label, color=color)
        # Add numbers above bars
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f"{height}",
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize=14)

    ax.set_xticks(x)
    ax.set_xticklabels(categories, rotation=20, fontsize=14)
    ax.set_ylabel("PromptOps Pass Rate (%)", fontsize=14)
    ax.set_title(title, fontsize=14)
    ax.legend(fontsize=14)
    ax.tick_params(axis='y', labelsize=14)
    plt.ylim(70, 105)
    plt.show()

# Replot GPT-4o PromptOps with 14 pt font size
plot_with_numbers(with_context_gpt, "QA (With context) - GPT-4o PromptOps")
plot_with_numbers(without_context_gpt, "QA (Without context) - GPT-4o PromptOps")
```
